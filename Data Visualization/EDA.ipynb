1. **What is EDA?**

EDA is an approach of exploring the data in a way where we are able to gather insights from the data.

- These insights may include the patterns that best explain the distribution of the data points, peakness and other necessary information like relationship between each of the variables in the data

- We are able to identify the underlying patterns which we may not be able to identify just by looking at it. We are able to remove the redundancies like null values and outliers.

* The **main objective** of EDA is to understand which kind of solution or if we are using machine learning approach we are going to use to solve the problem.



**2.Importance of Exploratory Data Analysi**

- Helps in uncovering underlying patterns and trends in the data that are not visible by just looking at the data
- Highlights relationship between data points to gather insights
- Helps restructure the data for modeling
- Helps in removing redundancies from the data
