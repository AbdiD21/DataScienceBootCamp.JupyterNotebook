## line plot

```python
x = np.arange(1,10,0.1)
y1 = 2*x+5
y2 = 3*x+5

plt.subplot(1,2,1) # one row and two column. 1 in the last is we are activating 1st subplot
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)

plt.show()
```

## bar plot

***The difference between bar plot and histogram***

- ***bar plot*** is used for categorial values
- ***histogram*** is used for numerical values

```python
# bar plot

fruit = {'apple':30,'mango':45,'banana':10}
print(fruit)
#next we will extract individual keys and values
names = list(fruit.keys())  # extract keys and put it in the form of list
quantity = list(fruit.values()) # extract values and put it in the form of list


names, quantity  # this prints individual names and quantity elements
```

```python
plt.bar(names,quantity)
plt.show()
```

```python
plt.bar(names,quantity)
plt.title("distribution of fruits ")
plt.xlabel("fruits")
plt.ylabel("quantity ")
plt.show()
```

```python
plt.bar(names,quantity, color = "orange")
plt.title("distribution of fruits ")
plt.xlabel("fruits")
plt.ylabel("quantity ")
plt.show()
```

```python
# to show it in the horizontal 
plt.barh(names,quantity, color = "orange")
plt.title("distribution of fruits ")
plt.xlabel("fruits")
plt.ylabel("quantity ")
plt.show()
```

## scatter plot

```python
# scatter plot - this compormise the relationship
x=[10,20,30,40,50,60,70,80]
a = [1,2,3,4,5,6,7,8]

plt.scatter(x,a)
plt.show()
```

```python
# scatter plot - this compormise the relationship
x=[10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,6,7,8,9]
b = [5,6,7,8,3,2,1,4,9]
plt.scatter(x,a)
plt.scatter(x,b)
plt.show()
```

```python
# customizing - scatter plot
x=[10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,4,3,2,9]
b = [5,6,7,8,3,2,1,4,9]
plt.scatter(x,a)
plt.scatter(x,b)
plt.title('x vs a&b')
plt.xlabel("x-axis")
plt.xlabel("y-axis")
plt.show()
```

```python
# customizing - scatter plot
x=[10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,4,3,2,9]
b = [5,6,7,8,3,2,1,4,9]
plt.scatter(x,a,s=100)# we added the size
plt.scatter(x,b, s=500)
plt.title('x vs a&b')
plt.xlabel("x-axis")
plt.xlabel("y-axis")
plt.show()
```

```python
# customizing - scatter plot
x=[10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,4,3,2,9]
b = [5,6,7,8,3,2,1,4,9]
plt.scatter(x,a,s=100,marker="3")# we changed the shape of this 
plt.scatter(x,b, s=500,marker="3")
plt.title('x vs a&b')
plt.xlabel("x-axis")
plt.xlabel("y-axis")
plt.show()
```

```python
# customizing - scatter plot
x=[10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,4,3,2,9]
b = [5,6,7,8,3,2,1,4,9]

plt.figure(figsize=(2,2))
plt.scatter(x,a,s=100,marker="3")# we changed the shape of this 
plt.scatter(x,b, s=500,marker="3")
plt.title('x vs a&b')
plt.xlabel("x-axis")
plt.xlabel("y-axis")
plt.show()
```

## violinplot
the difference between the boxplot and violinplot is only the heap difference

```python
# violinplot

one =[1,2,3,4,5,6,7,8,9]
two =[1,2,3,4,5,4,3,2,1]
three =[6,7,8,9,8,7,6,5,4]

data =list([one,two,three])

plt.violinplot(data)
plt.show()
```

```python
# violinplot

one =[1,2,3,4,5,6,7,8,9]
two =[1,2,3,4,5,4,3,2,1]
three =[6,7,8,9,8,7,6,5,4]

data =list([one,two,three])

plt.violinplot(data)
plt.grid(True)
plt.title("Distribution of data ")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# violinplot

one =[1,2,3,4,5,6,7,8,9]
two =[1,2,3,4,5,4,3,2,1]
three =[6,7,8,9,8,7,6,5,4]

data =list([one,two,three])

plt.violinplot(data,showmedians=True)# to add the medaian to this violinplot
plt.violinplot(data,showmeans=True)# to add the medaian to this violinplot

plt.grid(True)
plt.title("Distribution of data ")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

## piechart

```python
# pie chart

fruits =['apple', 'mango', 'orange','grapes']
quantity=[30,50,12,100]

plt.pie(quantity,labels=fruits)
plt.show()
```

```python
# pie chart

fruits =['apple', 'mango', 'orange','grapes']
quantity=[30,50,12,100]

plt.pie(quantity,labels=fruits,autopct='%0.1f%%')# here we added the percentage
plt.show()
```

```python
# pie chart

fruits =['apple', 'mango', 'orange','grapes']
quantity=[30,50,12,100]

plt.pie(quantity,labels=fruits,autopct='%0.1f%%',shadow=True)# here we added the shaddow
plt.show()
```

```python
# pie chart

fruits =['apple', 'mango', 'orange','grapes']
quantity=[30,50,12,100]

plt.pie(quantity,labels=fruits,autopct='%0.1f%%',shadow=True,explode=(0,0,0,0.1))# to highlight slice
# (0,0,0,0.1) means that the other three willnot be sliced. 0.1 is the amount the left is sliced
# plt.pie(quantity,labels=fruits,autopct='%0.1f%%',shadow=True,explode=(0,0,0,1))# to highlight slice
plt.show()
```

## pile chart

```python
# pie chart

fruits =['apple', 'mango', 'orange','grapes']
quantity=[30,50,12,100]

pie1=plt.pie(quantity,labels=fruits,autopct='%0.1f%%')
pie2=plt.pie([5],colors='white')
plt.show()
```

```python
# pie chart

fruits =['apple', 'mango', 'orange','grapes']
quantity=[30,50,12,100]

pie1=plt.pie(quantity,labels=fruits,autopct='%0.1f%%',radius=1)

pie2=plt.pie([5],colors='white',radius=1)
plt.show()
```
